Given you have been provided with the past 3 years of historical weekly sales data for 45 Walmart stores located in different regions and each store contains many departments.
1. How can you utilize these data to predict the next year sales for each department in each store ?
2. In the historical weekly sales data, selected holiday markdown events are included. These markdowns are known to affect sales, but how to predict which departments are affected by these markdowns?
3. And to what extent do these markdowns impact weekly sales?

This jupyter notebook demonstrate how I approach challenge questions from Walmart Recruiting competition.

Define evaluation function according to competition rules.

```python
def WMAE(y_test, y_predicted, x_test):
    y_test = y_test.to_numpy()
 
    W = x_test.IsHoliday.apply(lambda x: 5 if x else 1)
    W = W.to_numpy()
    return (1/np.sum(W))*np.sum(W*np.abs(y_test-y_predicted))
 
```

```python
wame = WMAE(y_test,test_pred, x_test )
wame
```

```python
def grid_search_RF(n_estimators, max_depth, X_train, Y_train):
    cv = pd.DataFrame(index = [0], columns = ['Max_Depth','Estimators','r2_score','WMAE'])
    for estimator in n_estimators:
        for depth in max_depth:
            wmaes_cv = []
            r2_score_cv = []
            for i in range(5):                 
                x_train, x_test, y_train, y_test = train_test_split(X_train, Y_train, test_size=0.3)
                RF = RandomForestRegressor(n_estimators=estimator, max_depth=depth)
                RF.fit(x_train, y_train)
                pred = RF.predict(x_test)
                r2_score_cv.append(r2_score(y_test,pred))
                wmaes_cv.append(WMAE(y_test, pred, x_test))
            print('n_estimators:', estimator, ', max_depth:', depth)
            print('r2_score', np.mean(r2_score_cv))
            print('WMAE:', np.mean(wmaes_cv))
            cv.loc[cv.index.max()+1] = [depth, estimator, np.mean(r2_score_cv), np.mean(wmaes_cv)]
    return cv[1:] 
```

```python
n_estimators = [10]
max_depth = [25,28,32]

grid_search_RF(n_estimators, max_depth, x_train, y_train)
```

```python
bestForest=RandomForestRegressor(n_estimators=50, max_depth=20)
bestForest.fit(x_train, y_train)
```

```python
test_pred=bestForest.predict(x_test)
r2_score(y_test,test_pred)
```
